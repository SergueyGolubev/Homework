```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

from wordcloud import WordCloud

import nltk
from nltk import word_tokenize, PorterStemmer
from nltk.corpus import stopwords

%matplotlib inline  
```

First we load emails and initialize a wordcloud instmance

```python
emails = pd.read_csv("hillary-clinton-emails/Emails.csv")
```

```python
wc = WordCloud(background_color="white")
```

According to the data schema from kaggle we should use 'ExtractedBodyText' as "Attempt to only pull out the text in the body that the email sender wrote"

```python
text = ' '.join(emails['ExtractedBodyText'].dropna())
```

```python
wc.generate(text)
plt.imshow(wc)
```

We see many additional verbs and stopwords in this result. Now we can try to implement a standard pipline where we perform tokenization, remove stopwords and apply steaming.

```python
porter = nltk.PorterStemmer()

tokens = word_tokenize(text)
# text = nltk.Text(tokens)
words = [w.lower() for w in tokens]
wordswostop = [w for w in words if w not in stopwords.words('english')]
wordsclean = [porter.stem(w) for w in wordswostop]
```

We can compare number of tokens before and after removing stopwords

```python
len(tokens), len(wordswostop)
```

```python
textclean = ' '.join(wordsclean)
wc.generate(textclean)
plt.imshow(wc)
```

As we can see we were able to remove some noisy stopwords. We also lost the endings. Overall we cannot say the results are sugnificant better than before
